**Research Questions:**

Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.
Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)


```python
# Filter data for relevant designations and fresh graduates
relevant_designations = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
fresh_graduates = df[(df['Designation'].isin(relevant_designations)) & (df['GraduationYear'] == 2018)] # Update with actual graduation year

# Calculate the maximum salary for fresh graduates in these designations
max_salary = fresh_graduates['Salary'].max()

# Create a histogram of salaries
plt.figure(figsize=(8, 6))
sns.histplot(fresh_graduates['Salary'], bins=20, color='skyblue', edgecolor='black')
plt.title('Salary Distribution of Fresh Graduates in Relevant Designations')
plt.xlabel('Salary (Lakhs)')
plt.ylabel('Frequency')

# Add a vertical line to represent the claimed salary range
plt.axvline(2.5, color='red', linestyle='--', label='Claim: 2.5 Lakhs')
plt.axvline(3.0, color='red', linestyle='--', label='Claim: 3.0 Lakhs')

# Add a legend
plt.legend()

# Display the plot
plt.show()

# Print the maximum salary and comparison with the claim
print(f"Maximum salary for fresh graduates in relevant designations: {max_salary}")
claim_range = (250000, 300000)  # 2.5-3 lakhs
if claim_range[0] <= max_salary <= claim_range[1]:
    print("The claim that fresh graduates earn upto 2.5-3 lakhs is supported by the data.")
else:
    print("The claim that fresh graduates earn upto 2.5-3 lakhs is not supported by the data.")
```

Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)


```python
from scipy.stats import chi2_contingency
# Assuming 'df' is your DataFrame with 'Gender' and 'Specialization' columns
gen_spe_tab = df[df['Specialization'] != 'Other']
# Create a contingency table
cont_tab = pd.crosstab(index=gen_spe_tab['Gender'], columns=gen_spe_tab['Specialization'])
# Perform the Chi-squared test
chi2, p, dof, expected = chi2_contingency(cont_tab)
# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
# Interpret the results
alpha = 0.05  # Significance level
if p < alpha:
    print("There is a significant relationship between gender and specialization.")
else:
    print("There is no significant relationship between gender and specialization.")
```
